## **Data Preprocessing KMP** (25-Nov-2023)

**Process :**
1. Import data using `!wget`
2. Read excel data using pandas library
3. Get the `"CODE"` coloumns for processing data
4. Using `data_cleaning` function
5. Get new data into new variable
6. Using pd.concat to get final data and split it into `final_data` and `duplicated_data`
7. If no duplicated data export data frame to excel using to_excel

Ikhwan Data usage:
* Cinsub_4_Fakultas Lain Ikhwan.xls

Akhwat Data usage:
* Cinsub_4_Fakultas Lain Akhwat.xls
* Cinsub_4_Fakultas Lain Akhwat_1.xls
* Cinsub_4_Fakultas Lain Akhwat_2.xls
* Cinsub_4_Fakultas Lain Akhwat_3.xls


```python
import pandas as pd

# Read dataset
data_ikhwan_mix = pd.read_excel("Cinsub_4_Fakultas Lain Ikhwan.xls")
data_akhwat_mix = pd.read_excel("Cinsub_4_Fakultas Lain Akhwat.xls")
data_akhwat_mix_1 = pd.read_excel("Cinsub_4_Fakultas Lain Akhwat_1.xls")
data_akhwat_mix_2 = pd.read_excel("Cinsub_4_Fakultas Lain Akhwat_2.xls")
data_akhwat_mix_3 = pd.read_excel("Cinsub_4_Fakultas Lain Akhwat_3.xls")


df_ikhwan = pd.DataFrame(data_ikhwan_mix)
df_akhwat = pd.DataFrame(data_akhwat_mix)
df_akhwat_1 = pd.DataFrame(data_akhwat_mix_1)
df_akhwat_2 = pd.DataFrame(data_akhwat_mix_2)
df_akhwat_3 = pd.DataFrame(data_akhwat_mix_3)


df_ikhwan
```

```python
df_code_ikhwan = df_ikhwan[0:]["CODE"]
df_code_akhwat = df_akhwat[0:]["CODE"]
df_code_akhwat_1 = df_akhwat_1[0:]["CODE"]
df_code_akhwat_2 = df_akhwat_2[0:]["CODE"]
df_code_akhwat_3 = df_akhwat_3[0:]["CODE"]

df_code_ikhwan
```

```python
def data_cleaning(data_frame):
  cleaning_data = []
  cleaning_data_min_3 = []
  cleaning_data_1 = []

  for i in data_frame.to_list():
    element = i.split(';\n')
    if len(element) == 3:
      cleaning_data.append(element)
    elif len(element) < 3:
      cleaning_data_min_3.append(element)
    else:
      cleaning_data_1.append([i])

  return cleaning_data, cleaning_data_min_3, cleaning_data_1
```

```python
data_cleaned_ikhwan = data_cleaning(df_code_ikhwan)
data_cleaned_akhwat = data_cleaning(df_code_akhwat)
data_cleaned_akhwat_1 = data_cleaning(df_code_akhwat_1)
data_cleaned_akhwat_2 = data_cleaning(df_code_akhwat_2)
data_cleaned_akhwat_3 = data_cleaning(df_code_akhwat_3)

data_cleaned_ikhwan
```

```python
# Ikhwan
clean_data_ikhwan = data_cleaned_ikhwan[0]
clean_data_ikhwan_min_3 = data_cleaned_ikhwan[2]
clean_data_ikhwan_1 = data_cleaned_ikhwan[1]

# Akhwat
clean_data_akhwat = data_cleaned_akhwat[0]
clean_data_akhwat_min_3 = data_cleaned_akhwat[2]
clean_data_akhwat_1 = data_cleaned_akhwat[1]

# Akhwat_1
clean_data_akhwat_1 = data_cleaned_akhwat_1[0]
clean_data_akhwat_min_3_1 = data_cleaned_akhwat_1[2]
clean_data_akhwat_1_1 = data_cleaned_akhwat_1[1]

# Akhwat_2
clean_data_akhwat_2 = data_cleaned_akhwat_2[0]
clean_data_akhwat_min_3_2 = data_cleaned_akhwat_2[2]
clean_data_akhwat_1_2 = data_cleaned_akhwat_2[2]

# Akhwat_3
clean_data_akhwat_3 = data_cleaned_akhwat_3[0]
clean_data_akhwat_min_3_3 = data_cleaned_akhwat_3[2]
clean_data_akhwat_1_3 = data_cleaned_akhwat_3[2]


clean_data_ikhwan_1
```

```python
# Looping to clean single data
def cleaning_data_single(data):
  clean_ide = []
  clean_detail = []
  for i in data:
    for x in i:
      if len(x) <= 10:
        clean_detail.append(x)
      else:
        clean_ide.append(x)
  return clean_detail

single_data_ikhwan = cleaning_data_single(clean_data_ikhwan_1)
single_data_akhwat = cleaning_data_single(clean_data_akhwat_1)
single_data_akhwat_1 = cleaning_data_single(clean_data_akhwat_1_1)
single_data_akhwat_2 = cleaning_data_single(clean_data_akhwat_1_2)
single_data_akhwat_3 = cleaning_data_single(clean_data_akhwat_1_3)

single_data_ikhwan
```

```python
new_data_ikhwan = pd.DataFrame(clean_data_ikhwan, columns=("NIM", "TTL", "Alamat"))
new_data_akhwat = pd.DataFrame(clean_data_akhwat, columns=("NIM", "TTL", "Alamat"))
new_data_akhwat_1 = pd.DataFrame(clean_data_akhwat_1, columns=("NIM", "TTL", "Alamat"))
new_data_akhwat_2 = pd.DataFrame(clean_data_akhwat_2, columns=("NIM", "TTL", "Alamat"))
new_data_akhwat_3 = pd.DataFrame(clean_data_akhwat_3, columns=("NIM", "TTL", "Alamat"))

new_single_data_ikhwan_1 = pd.DataFrame(single_data_ikhwan, columns=["NIM"])
new_single_data_akhwat_1 = pd.DataFrame(single_data_akhwat, columns=["NIM"])
new_single_data_akhwat_1_1 = pd.DataFrame(single_data_akhwat_1, columns=["NIM"])
new_single_data_akhwat_1_2 = pd.DataFrame(single_data_akhwat_2, columns=["NIM"])
new_single_data_akhwat_1_3 = pd.DataFrame(single_data_akhwat_3, columns=["NIM"])

new_single_data_akhwat_1
```

```python
combinated_data_akhwat_clean = pd.concat([new_data_akhwat, new_data_akhwat_1, new_data_akhwat_2, new_data_akhwat_3], ignore_index=True)
combinated_data_akhwat_dirty = pd.concat([new_single_data_akhwat_1, new_single_data_akhwat_1_1, new_single_data_akhwat_1_2, new_single_data_akhwat_1_3], ignore_index=True)

combinated_data_ikhwan_clean = new_data_ikhwan
combinated_data_ikhwan_dirty = new_single_data_ikhwan_1

combinated_data_ikhwan_dirty
```

```python
# Filtering data based on NIM Alphabet
sorted_df_ikhwan_clean = combinated_data_ikhwan_clean.sort_values(by="NIM").drop_duplicates().reset_index(drop=True)
sorted_df_ikhwan_dirty = combinated_data_ikhwan_dirty.sort_values(by="NIM").drop_duplicates().reset_index(drop=True)
sorted_df_akhwat_clean = combinated_data_akhwat_clean.sort_values(by="NIM").drop_duplicates().reset_index(drop=True)
sorted_df_akhwat_dirty = combinated_data_akhwat_dirty.sort_values(by="NIM").drop_duplicates().reset_index(drop=True)

sorted_df_ikhwan_clean
```

```python
# Filtering data based FKI Members
filtered_df_ikhwan_fki_clean = sorted_df_ikhwan_clean[sorted_df_ikhwan_clean['NIM'].str.startswith('L')].drop_duplicates().reset_index(drop=True)
filtered_df_ikhwan_fki_dirty = sorted_df_ikhwan_dirty[sorted_df_ikhwan_dirty['NIM'].str.startswith('L')].drop_duplicates().reset_index(drop=True)
filtered_df_akhwat_fki_clean = sorted_df_akhwat_clean[sorted_df_akhwat_clean['NIM'].str.startswith('L')].drop_duplicates().reset_index(drop=True)
filtered_df_akhwat_fki_dirty = sorted_df_akhwat_dirty[sorted_df_akhwat_dirty['NIM'].str.startswith('L')].drop_duplicates().reset_index(drop=True)


filtered_df_ikhwan_fki_clean
```

```python
# Save Data to Local
sorted_df_ikhwan_clean.to_excel("Cinsub_4_(25-Nov-2023)_Fakultas Lain Ikhwan_Data-Bersih.xlsx")
sorted_df_ikhwan_dirty.to_excel("Cinsub_4_(25-Nov-2023)_Fakultas Lain Ikhwan_Data-Kotor.xlsx")
sorted_df_akhwat_clean.to_excel("Cinsub_4_(25-Nov-2023)_Fakultas Lain Akhwat_Data-Bersih.xlsx")
sorted_df_akhwat_dirty.to_excel("Cinsub_4_(25-Nov-2023)_Fakultas Lain Akhwat_Data-Kotor.xlsx")
```

```python
# Save Data to Local
filtered_df_ikhwan_fki_clean.to_excel("Cinsub_4_(25-Nov-2023)_FKI-Ikhwan_Data-Bersih.xlsx")
filtered_df_ikhwan_fki_dirty.to_excel("Cinsub_4_(25-Nov-2023)_FKI-Ikhwan_Data-Kotor.xlsx")
filtered_df_akhwat_fki_clean.to_excel("Cinsub_4_(25-Nov-2023)_FKI-Akhwat_Data-Bersih.xlsx")
filtered_df_akhwat_fki_dirty.to_excel("Cinsub_4_(25-Nov-2023)_FKI-Akhwat_Data-Kotor.xlsx")
```
