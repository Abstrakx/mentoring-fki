## Data Preprocessing KM-FKI

**Process :**
1. Import data using `!wget`
2. Read excel data using pandas library
3. Get the `"CODE"` coloumns for processing data
4. Using `data_cleaning` function
5. Get new data into new variable
6. Using pd.concat to get final data and split it into `final_data` and `duplicated_data`
7. If no duplicated data export data frame to excel using to_excel

Ikhwan Data usage:
* Cinsub_18_Nov_2023_FKI_Ikhwan_Alif.xls
* Cinsub_18_Nov_2023_FKI_Ikhwan_Iqbal.xls
* Cinsub_18_Nov_2023_FKI_Ikhwan_Dena.xls

Akhwat Data usage:
* Cinsub_18_Nov_2023_FKI_Akhwat_Afifah.xls
* Cinsub_18_Nov_2023_FKI_Akhwat_Nasyawa.xls

```python
import pandas as pd

# Read dataset
data_afifah = pd.read_excel("Cinsub_18_Nov_2023_FKI_Akhwat_Afifah.xls")
data_nasyawa = pd.read_excel("Cinsub_18_Nov_2023_FKI_Akhwat_Nasyawa.xls")
data_iqbal = pd.read_excel("Cinsub_18_Nov_2023_FKI_Ikhwan_Iqbal.xls")
data_alif = pd.read_excel("Cinsub_18_Nov_2023_FKI_Ikhwan_Alif.xls")
data_dena = pd.read_excel("Cinsub_18_Nov_2023_FKI_Ikhwan_Dena.xls")

df_afifah = pd.DataFrame(data_afifah)
df_nasyawa = pd.DataFrame(data_nasyawa)
df_dena = pd.DataFrame(data_dena)
df_iqbal = pd.DataFrame(data_iqbal)
df_alif = pd.DataFrame(data_alif)

df_afifah
```

```python
df_code_afifah = df_afifah[0:]["CODE"]
df_code_nasyawa = df_nasyawa[0:]["CODE"]
df_code_iqbal = df_iqbal[0:]["CODE"]
df_code_dena = df_dena[0:]["CODE"]
df_code_alif = df_alif[0:]["CODE"]

df_code_afifah
```

```python
def data_cleaning(data_frame):
  cleaning_data = []

  for i in data_frame.to_list():
    element = i.split(';\n')
    if len(element) == 3:
      cleaning_data.append(element)
  return cleaning_data
```

```python
data_cleaned_afifah = data_cleaning(df_code_afifah)
data_cleaned_nasyawa = data_cleaning(df_code_nasyawa)
data_cleaned_dena = data_cleaning(df_code_dena)
data_cleaned_iqbal = data_cleaning(df_code_iqbal)
data_cleaned_alif = data_cleaning(df_code_alif)

data_cleaned_afifah
```

```python
new_data_afifah = pd.DataFrame(data_cleaned_afifah, columns=("NIM", "TTL", "Alamat"))
new_data_nasyawa = pd.DataFrame(data_cleaned_nasyawa, columns=("NIM", "TTL", "Alamat"))
new_data_dena = pd.DataFrame(data_cleaned_dena, columns=("NIM", "TTL", "Alamat"))
new_data_iqbal = pd.DataFrame(data_cleaned_iqbal, columns=("NIM", "TTL", "Alamat"))
new_data_alif = pd.DataFrame(data_cleaned_alif, columns=("NIM", "TTL", "Alamat"))

new_data_iqbal
```

```python
filtered_df_afifah = new_data_afifah[new_data_afifah['NIM'].str.startswith('L')]
filtered_df_nasyawa = new_data_nasyawa[new_data_nasyawa['NIM'].str.startswith('L')]
filtered_df_dena = new_data_dena[new_data_dena['NIM'].str.startswith('L')]
filtered_df_iqbal = new_data_iqbal[new_data_iqbal['NIM'].str.startswith('L')]
filtered_df_alif = new_data_alif[new_data_alif['NIM'].str.startswith('L')]

filtered_df_afifah
```

```python
combinated_data_ikhwan = pd.concat([filtered_df_alif, filtered_df_iqbal, filtered_df_dena], ignore_index=True)
combinated_data_akhwat = pd.concat([filtered_df_nasyawa, filtered_df_afifah], ignore_index=True)

final_data_akhwat = combinated_data_akhwat
duplicated_data_akhwat = combinated_data_akhwat[combinated_data_akhwat.duplicated()]
final_data_ikhwan = combinated_data_ikhwan
duplicated_data_ikhwan = combinated_data_ikhwan[combinated_data_ikhwan.duplicated()]

final_data_akhwat
```
