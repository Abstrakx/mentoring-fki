## Data Preprocessing Akhwat

**Process :**
1. Import data using `!wget`
2. Read excel data using pandas library
3. Get the `"CODE"` coloumns for processing data
4. Using `data_cleaning` function
5. Get new data into new variable
6. Using pd.concat to get final data and split it into `final_data` and `duplicated_data`
7. If no duplicated data export data frame to excel using to_excel

Data usage:
* Cinsub_18_Nov_2023_FKI_Akhwat_Afifah.xls
* Cinsub_18_Nov_2023_FKI_Akhwat_Nasyawa.xls

```python
# Get dataset from github repo (Github make data error)
!wget https://github.com/Abstrakx/mentoring-fki/blob/main/Data-Cinsub-18_Nov_2023/Cinsub_18_Nov_2023_FKI_Akhwat_Afifah.xls
!wget https://github.com/Abstrakx/mentoring-fki/blob/main/Data-Cinsub-18_Nov_2023/Cinsub_18_Nov_2023_FKI_Akhwat_Nasyawa.xls
```

```python
import pandas as pd

# Read dataset
data_afifah = pd.read_excel("Cinsub_9_Des_2023_FKI_Akhwat_Afifah.xls")
data_afifah_1 = pd.read_excel("Cinsub_9_Des_2023_FKI_Akhwat_Afifah_1.xls")
data_ami = pd.read_excel("Cinsub_9_Des_2023_FKI_Akhwat_Ami.xls")
data_iqbal = pd.read_excel("Cinsub_9_Des_2023_FKI_Ikhwan_Iqbal.xls")
data_iqbal_1 = pd.read_excel("Cinsub_9_Des_2023_FKI_Ikhwan_Iqbal_1.xls")

df_afifah = pd.DataFrame(data_afifah)
df_afifah_1 = pd.DataFrame(data_afifah_1)
df_ami = pd.DataFrame(data_ami)
df_iqbal = pd.DataFrame(data_iqbal)
df_iqbal_1 = pd.DataFrame(data_iqbal_1)

df_afifah
```

```python
df_code_afifah = df_afifah[0:]["CODE"]
df_code_afifah_1 = df_afifah_1[0:]["CODE"]
df_code_ami = df_ami[0:]["CODE"]
df_code_iqbal = df_iqbal[0:]["CODE"]
df_code_iqbal_1 = df_iqbal_1[0:]["CODE"]

df_code_ami
```

```python
def data_cleaning(data_frame):
  cleaning_data = []

  for i in data_frame.to_list():
    element = i.split(';\n')
    if len(element) == 3:
      cleaning_data.append(element)
  return cleaning_data
```

```python
data_cleaned_afifah = data_cleaning(df_code_afifah)
data_cleaned_afifah_1 = data_cleaning(df_code_afifah_1)
data_cleaned_ami = data_cleaning(df_code_ami)
data_cleaned_iqbal = data_cleaning(df_code_iqbal)
data_cleaned_iqbal_1 = data_cleaning(df_code_iqbal_1)
data_cleaned_afifah
```

```python
new_data_afifah = pd.DataFrame(data_cleaned_afifah, columns=("NIM", "TTL", "Alamat"))
new_data_afifah_1 = pd.DataFrame(data_cleaned_afifah_1, columns=("NIM", "TTL", "Alamat"))
new_data_ami = pd.DataFrame(data_cleaned_ami, columns=("NIM", "TTL", "Alamat"))
new_data_iqbal = pd.DataFrame(data_cleaned_iqbal, columns=("NIM", "TTL", "Alamat"))
new_data_iqbal_1 = pd.DataFrame(data_cleaned_iqbal_1, columns=("NIM", "TTL", "Alamat"))

new_data_iqbal
```

```python
filtered_df_afifah = new_data_afifah[new_data_afifah['NIM'].str.startswith('L')]
filtered_df_afifah_1 = new_data_afifah_1[new_data_afifah_1['NIM'].str.startswith('L')]
filtered_df_ami = new_data_ami[new_data_ami['NIM'].str.startswith('L')]
filtered_df_iqbal = new_data_iqbal[new_data_iqbal['NIM'].str.startswith('L')]
filtered_df_iqbal_1 = new_data_iqbal_1[new_data_iqbal_1['NIM'].str.startswith('L')]

filtered_df_afifah
```

```python
combinated_data_akhwat = pd.concat([filtered_df_afifah, filtered_df_afifah_1, filtered_df_ami], ignore_index=True)
combinated_data_ikhwan = pd.concat([filtered_df_iqbal, filtered_df_iqbal_1], ignore_index=True)

final_data_akhwat = combinated_data_akhwat
duplicated_data_akhwat = combinated_data_akhwat[combinated_data_akhwat.duplicated()]
final_data_ikhwan = combinated_data_ikhwan
duplicated_data_ikhwan = combinated_data_ikhwan[combinated_data_ikhwan.duplicated()]

final_data_akhwat
```
