## **Data Preprocessing KMP** (28-Oct-2023)

**Process :**
1. Import data using `!wget`
2. Read excel data using pandas library
3. Get the `"CODE"` coloumns for processing data
4. Using `data_cleaning` function
5. Get new data into new variable
6. Using pd.concat to get final data and split it into `final_data` and `duplicated_data`
7. If no duplicated data export data frame to excel using to_excel

Ikhwan Data usage:
* Cinsub_2_Fakultas Lain Ikhwan.xlsx

Akhwat Data usage:
* Cinsub_2_Fakultas Lain Akhwat.xlsx

```python
import pandas as pd

# Read dataset
data_ikhwan_mix = pd.read_excel("Cinsub_2_Fakultas Lain Ikhwan.xlsx")
data_akhwat_mix = pd.read_excel("Cinsub_2_Fakultas Lain Akhwat.xlsx")

df_ikhwan = pd.DataFrame(data_ikhwan_mix)
df_akhwat = pd.DataFrame(data_akhwat_mix)

df_ikhwan
```

```python
df_code_ikhwan = df_ikhwan[0:]["CODE"]
df_code_akhwat = df_akhwat[0:]["CODE"]

df_code_ikhwan
```

```python
def data_cleaning(data_frame):
  cleaning_data = []
  cleaning_data_min_3 = []
  cleaning_data_1 = []

  for i in data_frame.to_list():
    element = i.split(';\n')
    if len(element) == 3:
      cleaning_data.append(element)
    elif len(element) < 3:
      cleaning_data_min_3.append(element)
    else:
      cleaning_data_1.append([i])

  return cleaning_data, cleaning_data_min_3, cleaning_data_1
```

```python
data_cleaned_ikhwan = data_cleaning(df_code_ikhwan)
data_cleaned_akhwat = data_cleaning(df_code_akhwat)

data_cleaned_ikhwan
```

```python
# Ikhwan
clean_data_ikhwan = data_cleaned_ikhwan[0]
clean_data_ikhwan_min_3 = data_cleaned_ikhwan[2]
clean_data_ikhwan_1 = data_cleaned_ikhwan[1]

# Akhwat
clean_data_akhwat = data_cleaned_akhwat[0]
clean_data_akhwat_min_3 = data_cleaned_akhwat[2]
clean_data_akhwat_1 = data_cleaned_akhwat[1]

clean_data_ikhwan_1
```

```python
new_data_ikhwan = pd.DataFrame(clean_data_ikhwan, columns=("NIM", "TTL", "Alamat"))
new_data_akhwat = pd.DataFrame(clean_data_akhwat, columns=("NIM", "TTL", "Alamat"))
new_data_ikhwan_1 = pd.DataFrame(clean_data_ikhwan_1, columns=["NIM"])
new_data_akhwat_1 = pd.DataFrame(clean_data_akhwat_1, columns=["NIM"])

new_data_ikhwan
```

```python
sorted_df_ikhwan_clean = new_data_ikhwan.sort_values(by="NIM").reset_index(drop=True)
sorted_df_ikhwan_clean_1 = new_data_ikhwan_1.sort_values(by="NIM").reset_index(drop=True)
sorted_df_akhwat_clean = new_data_akhwat.sort_values(by="NIM").reset_index(drop=True)
sorted_df_akhwat_clean_1 = new_data_akhwat_1.sort_values(by="NIM").reset_index(drop=True)

sorted_df_akhwat_clean_1
```

```python
# Filtering data based on NIM Alphabet
filtered_df_ikhwan_fki = sorted_df_ikhwan_clean[sorted_df_ikhwan_clean['NIM'].str.startswith('L')].reset_index(drop=True)
filtered_df_ikhwan_fki_1 = sorted_df_ikhwan_clean_1[sorted_df_ikhwan_clean_1['NIM'].str.startswith('L')].reset_index(drop=True)
filtered_df_akhwat_fki = sorted_df_akhwat_clean[sorted_df_akhwat_clean['NIM'].str.startswith('L')].reset_index(drop=True)
filtered_df_akhwat_fki_1 = sorted_df_akhwat_clean_1[sorted_df_akhwat_clean_1['NIM'].str.startswith('L')].reset_index(drop=True)


filtered_df_akhwat_fki
```

```python
# Save Data to Local
sorted_df_ikhwan_clean.to_excel("Cinsub_2_(28-Okt-2023)_Fakultas Lain Ikhwan_Data-Bersih.xlsx")
sorted_df_ikhwan_clean_1.to_excel("Cinsub_2_(28-Okt-2023)_Fakultas Lain Ikhwan_Data-Kotor.xlsx")
sorted_df_akhwat_clean.to_excel("Cinsub_2_(28-Okt-2023)_Fakultas Lain Akhwat_Data-Bersih.xlsx")
sorted_df_akhwat_clean_1.to_excel("Cinsub_2_(28-Okt-2023)_Fakultas Lain Akhwat_Data-Kotor.xlsx")
```

```python
# Save Data to Local
filtered_df_ikhwan_fki.to_excel("Cinsub_2_(28-Okt-2023)_FKI-Ikhwan_Data-Bersih.xlsx")
filtered_df_ikhwan_fki_1.to_excel("Cinsub_2_(28-Okt-2023)_FKI-Ikhwan_Data-Kotor.xlsx")
filtered_df_akhwat_fki.to_excel("Cinsub_2_(28-Okt-2023)_FKI-Akhwat_Data-Bersih.xlsx")
filtered_df_akhwat_fki_1.to_excel("Cinsub_2_(28-Okt-2023)_FKI-Akhwat_Data-Kotor.xlsx")
```
